## 探索式資料分析 EDA

```python
import pandas as pd

# Load the dataset
file_path = 'E-commerce Customer Behavior - Sheet1.csv'
ecommerce_data = pd.read_csv(file_path)

ecommerce_data
```

檢視不同性別、年齡、會員種類和平均總花費之間的關係

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the visualisation style
sns.set(style="whitegrid")

# Customer Behavior Analysis by Gender, Age Group, and Membership Type

# Creating age groups for better analysis
bins = [25, 33, 38, 44]
labels = ['26-32', '33-37', '38-43']
ecommerce_data['Age Group'] = pd.cut(ecommerce_data['Age'], bins=bins, labels=labels, right=False)

# Analyzing average spend by Gender
avg_spend_gender = ecommerce_data.groupby('Gender')['Total Spend'].mean().reset_index()

# Analyzing average spend by Age Group
avg_spend_age_group = ecommerce_data.groupby('Age Group')['Total Spend'].mean().reset_index()

# Analyzing average spend by Membership Type
avg_spend_membership = ecommerce_data.groupby('Membership Type')['Total Spend'].mean().reset_index()

# Creating subplots
fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# Plotting
sns.barplot(x='Gender', y='Total Spend', data=avg_spend_gender, ax=ax[0])
ax[0].set_title('Average Spend by Gender')

sns.barplot(x='Age Group', y='Total Spend', data=avg_spend_age_group, ax=ax[1])
ax[1].set_title('Average Spend by Age Group')

sns.barplot(x='Membership Type', y='Total Spend', data=avg_spend_membership, ax=ax[2])
ax[2].set_title('Average Spend by Membership Type')

plt.tight_layout()
plt.show()
```

對數值型變數的進行摘要統計分析，並繪製出分佈圖、添加核密度估計（KDE）。

`核心密度估計`：它是一種平滑技術，用於展示數據的分布情況，提供比傳統直方圖更平滑、更連續的數據分布視圖。

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting up the aesthetic style of the plots
sns.set(style="whitegrid")

# Summary statistics for numerical columns
summary_stats = ecommerce_data.describe()

# Plotting distributions of key numerical columns
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Distribution of Age
sns.histplot(ecommerce_data['Age'], bins=30, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Age')

# Distribution of Total Spend
sns.histplot(ecommerce_data['Total Spend'], bins=30, kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Total Spend')

# Distribution of Items Purchased
sns.histplot(ecommerce_data['Items Purchased'], bins=30, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Items Purchased')

# Distribution of Average Rating
sns.histplot(ecommerce_data['Average Rating'], bins=30, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Average Rating')

# Distribution of Days Since Last Purchase
sns.histplot(ecommerce_data['Days Since Last Purchase'], bins=30, kde=True, ax=axes[2, 0])
axes[2, 0].set_title('Days Since Last Purchase')

# Close the empty subplot
axes[2, 1].axis('off')

plt.tight_layout()
plt.show()

summary_stats
```

進一步將電商平台數據中各變量間的關係和趨勢做視覺化呈現。

```python
# Analyzing relationships and trends among various variables

# Setting up the aesthetic style of the plots
sns.set(style="whitegrid")

fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Total Spend vs Age
sns.scatterplot(data=ecommerce_data, x='Age', y='Total Spend', hue='Gender', ax=axes[0, 0])
axes[0, 0].set_title('Total Spend vs Age')

# Total Spend vs Membership Type
sns.boxplot(data=ecommerce_data, x='Membership Type', y='Total Spend', ax=axes[0, 1])
axes[0, 1].set_title('Total Spend vs Membership Type')

# Satisfaction Level vs Total Spend
sns.boxplot(data=ecommerce_data, x='Satisfaction Level', y='Total Spend', ax=axes[1, 0])
axes[1, 0].set_title('Satisfaction Level vs Total Spend')

# Average Rating vs Days Since Last Purchase
sns.scatterplot(data=ecommerce_data, x='Days Since Last Purchase', y='Average Rating', ax=axes[1, 1])
axes[1, 1].set_title('Average Rating vs Days Since Last Purchase')

# Distribution of Satisfaction Level
sns.countplot(data=ecommerce_data, x='Satisfaction Level', ax=axes[2, 0])
axes[2, 0].set_title('Distribution of Satisfaction Levels')

# Total Spend by City
sns.barplot(data=ecommerce_data, x='City', y='Total Spend', estimator=sum, ci=None, ax=axes[2, 1])
axes[2, 1].set_title('Total Spend by City')

plt.tight_layout()
plt.show()
```

### 滿意度分析
找出哪些因素最直接影響滿意度，透過 correlation matrix 也可觀察變量間的關係。

```python
# Investigate null value
missing_values_new = ecommerce_data.isnull().sum()
missing_values_new
```

```python
# Calculating the correlation matrix for numerical variables
correlation_matrix = ecommerce_data.corr()

correlation_matrix
```

```python
# Analyzing the relationship between satisfaction level and other factors

# Converting Satisfaction Level, Membership Type into an ordinal category for analysis

satisfaction_mapping = {'Unsatisfied': 0, 'Neutral': 1, 'Satisfied': 2}
ecommerce_data['Satisfaction Level Ordinal'] = ecommerce_data['Satisfaction Level'].map(satisfaction_mapping)

membership_mapping = {'Bronze': 0, 'Silver': 1, 'Gold': 2}
ecommerce_data['Membership Ordinal'] = ecommerce_data['Membership Type'].map(membership_mapping)

# Correlation matrix
correlation_factors = ['Total Spend', 'Items Purchased', 'Average Rating', 'Days Since Last Purchase', 'Discount Applied', 'Satisfaction Level Ordinal', 'Membership Ordinal']
correlation_data = ecommerce_data[correlation_factors].corr()

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap between All Factors')
plt.show()
```

```python
# Identifying the factor with the highest correlation to Average Rating
correlation_with_rating = correlation_data['Average Rating'].drop('Average Rating')  # drop self-correlation
most_significant_factor = correlation_with_rating.abs().idxmax()
correlation_value = correlation_with_rating[most_significant_factor]

most_significant_factor, correlation_value
```

### 客戶分群

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Selecting features for clustering
cluster_features = ['Age', 'Total Spend', 'Items Purchased', 'Membership Type']
data_clustering = ecommerce_data[cluster_features]

# One-hot encoding for 'Membership Type'
encoder = OneHotEncoder(sparse=False)
encoded = encoder.fit_transform(data_clustering[['Membership Type']])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())

# Concatenating the one-hot encoded columns with the original data
data_clustering_final = pd.concat([data_clustering.drop(['Membership Type'], axis=1), encoded_df], axis=1)
```

```python
data_clustering_final
```

```python
# Standardizing the data
scaler = StandardScaler()
data_clustering_scaled = scaler.fit_transform(data_clustering_final)

# Finding the optimal number of clusters using the Elbow Method
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(data_clustering_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Determining Best K Value')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.annotate('Elbow', xy=(3, inertia[2]), xytext=(4, 350),
             arrowprops=dict(arrowstyle='->', lw=1.5, color='red'))
plt.grid(True)
plt.show()
```

根據肘部法則的結果，選擇將資料分成 3 群。

```python
# Applying K-means with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
ecommerce_data['Cluster'] = kmeans.fit_predict(data_clustering_scaled)

# Analyzing the clusters
cluster_analysis = ecommerce_data.groupby('Cluster')[cluster_features].mean()
cluster_analysis['Count'] = ecommerce_data['Cluster'].value_counts()

membership_count_in_clusters = ecommerce_data.groupby(['Cluster', 'Membership Type']).size().unstack(fill_value=0)

final_cluster_analysis = cluster_analysis.join(membership_count_in_clusters)
final_cluster_analysis
```

```python
grouped_stats = ecommerce_data.groupby('Membership Type').mean()
grouped_stats_sorted = grouped_stats.sort_values(by='Total Spend', ascending=False)
grouped_stats_sorted[['Cluster', 'Age', 'Total Spend', 'Items Purchased', 'Average Rating', 'Discount Applied', 'Days Since Last Purchase']]
```
